#### modules and given values

```python
import numpy as np
import matplotlib.pyplot as plt

V1, V2 = 100, 0       # boundary values at (1) top and (2) bottom and left
a = 1                 # length of sides of the square
# precision = 0.0001
```

#### initial potential values
Returns a square $L+2$ by $L+2$ matrix, with $L^2$ internal points and the boundary points surrounding them.
It was based on an analytic guess. The values were assumed to vary linearly along the lines making $45^\circ$ with the y-axis: from 0 V to (1) 100 V in the left-upper half and (2) linearly changing voltage values on the open boundary. This required fewer iterations than using most other values (random, zeros).

```python
def V_init(L):
    V0 = np.empty([L+2, L+2], dtype=float)
    
    # boundary values
    # 100 V on top; 0 V at left and bottom; linear at right
    for j in range(L+2):
        V0[0][j] = V1
        V0[L+1][j] = V2
        
    for i in range(1, L+1):
        V0[i][0] = V2
        V0[i][L+1] = (1 - i/(L+1)) * V1
        
    # interior points
    for j in range(1, L+1):
        for i in range(1, L+1):
            if i+j <= L+1:
                V0[i][j] = (j / (i+j)) * V1
            else:
                V0[i][j] = ((L+1-i) / (2*(L+1)-i-j)) * V0[i+j-L-1][L+1]
            # V0[i][j] = 0.0
    return V0
```

#### average value from neighboring points

```python
def avg(V, i, j):
    return (V[i-1][j-1] + V[i-1, j+1] + V[i+1][j-1] + V[i+1][j+1]) / 4
```

#### sum of change in potential at every point between successive iterations

```python
def diff(V, temp, L):
    d = 0.0
    for i in range(L+2):
        for j in range(L+2): 
            d += abs(V[i][j] - temp[i][j])   
    return d
```

#### numerical solution using Jacobs, Gauss-Seidel and SOR methods
For input, the type of method is passed as string of the name ("Jacobi", "Gauss-Seidel", or "SOR") along with the number of internal points $L$ and the precision $p$ required for solution. It returns the solution matrix, the number of iterations required, and the name of the method.

The values were updated in temp matrix, which was used in Gauss-Seidel and SOR methods to calculate the average value from neighboring points, while Jacobi method used the matrix from previous iteration. The optimization parameter $w$ was tuned in SOR method as 
$$ w = \frac{2}{1+\pi L}. $$

$w=1$ in the other two methods.

```python
def soln(method, L, p=0.0001):
    V = V_init(L)
    temp = np.copy(V)
    
    n = 0
    d = p + 1.0      # difference between successive iterations; set d > p to enter loop
    while d > p:
        if method == "Jacobi":
            avg_arr, w = V, 1
        elif method == "Gauss-Seidel":
            avg_arr, w = temp, 1
        elif method == "SOR":
            avg_arr, w = temp, 2 / (1 + np.pi / L)
        
        # iterate along the rows and update open boundary value at the end of each row
        # interior points
        for i in range(1, L+1):
            for j in range(1, L+1):
                temp[i][j] = w * avg(avg_arr, i, j) + (1 - w) * V[i][j]

            # open boundary
            temp[i][L+1] = temp[i][L]

        d = diff(V, temp, L)
        V = np.copy(temp)
        n += 1
    return V, n, method
```

#### part a) and b)
Solutions using different methods. The grid spacings were kept different smaller grid spacing in methods converging slower would increase the runtime significantly.

```python
Va = soln("Jacobi", 20)
Vb = soln("Gauss-Seidel", 30)
Vc = soln("SOR", 40)

plot(Va)
plot(Vb)
plot(Vc)
```

#### part c)
$L$ is used instead of step size $h$.

```python
L_arr = [2, 4, 6, 8, 10, 12, 16]
num = len(L_arr)

# a, b, c refer to Jacobi, Gauss-Seidel and SOR respectively
a, b, c = [], [], []       # lists to contain solutions
na, nb, nc = [], [], []    # lists to contain number of iterations
da, db, dc = [], [], []    # lists to contain difference from better solution

# calculates the average potential for a solution
def avg_pot(V_obj):
    V = np.array(V_obj)    # if not array
    L = V.shape[0]-2
    sum_potential = 0.0
    
    for i in range(1, L+1):
        for j in range(1, L+1):
            sum_potential += V[i][j]
    return sum_potential / (L**2)

# "best" SOR solution with L = 60
v0 = avg_pot(soln("SOR", 60)[0])

for i in range(num):
    a.append(soln("Jacobi", L_arr[i]))
    b.append(soln("Gauss-Seidel", L_arr[i]))
    c.append(soln("SOR", L_arr[i]))
    
    na.append(a[i][1])
    nb.append(b[i][1])
    nc.append(c[i][1])

    da.append(avg_pot(a[i][0]) - v0)
    db.append(avg_pot(b[i][0]) - v0)
    dc.append(avg_pot(c[i][0]) - v0)
    
# effect of L (or h) on convergence speed
plt.plot(L_arr, na, label="Jacobi")
plt.plot(L_arr, nb, label="Gauss-Seidel")
plt.plot(L_arr, nc, label="SOR")
plt.title('effect of L (or h) on convergence speed')
plt.xlim(0, L_arr[num-1] + 5)
plt.xlabel('L')
plt.ylabel('no. of iterations')
plt.legend()
plt.show()

# effect of h on accuracy
plt.plot(L_arr, va, label="Jacobi")
plt.plot(L_arr, vb, label="Gauss-Seidel")
plt.plot(L_arr, vc, label="SOR")
plt.title('effect of L (or h) on accuracy')
plt.xlim(0, L_arr[num-1] + 5)
plt.xlabel('L')
plt.ylabel('difference from best solution [V]')
plt.legend()
plt.show()
```

#### part d)
It can be concluded from the result in part c) that, as expected, SOR method > Gauss-Seidel > Jacobi in terms of converging speed. It can also be seen that the accuracy of the solution is independent of the method and seems to decrease exponentially with L. Hence, again, SOR method > Gauss-Seidel > Jacobi in terms of better methods here.
