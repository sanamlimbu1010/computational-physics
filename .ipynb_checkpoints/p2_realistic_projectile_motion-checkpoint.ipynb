### project summary
The code is written for the general case in $n$ and initial velocity (magnitude $v_0$ and angle $\theta_0$ with the horizontal). <br> 
The cases for $n=1$ in part b) and $n=1.5$ and $n=2$ in part c) are implemented simply by passing respective values for $n$. <br>
Using the condition that the initial frictional force has to be equal for any case, the coefficient $k_b$ for $n_b$ can be obtained as 
    $$k_b = k_a v_0^{n_a-n_b},$$
where k_a = 0.8 for n_a = 1 and $v_0 = 22\, m/s$ are given. For part d) different values of $v_0$ will be be used, but the initial frictional force will be the same. <br>

### part a)
Using Newton's second law, we have
\begin{align}
    & m\frac{d^2 \vec r}{dt^2} = -mg \hat{y} - km|v|^n \frac{\vec{v}}{|v|} \\
    \implies \quad & \frac{d^2}{dt^2}(x \hat x + y \hat y) = -g \hat y -k|v|^{n-1}(v_x \hat x + v_y \hat y) \\
    \implies \quad & \frac{d^2 x}{dt^2} = -k|v|^{n-1}v_x \quad and \quad \frac{d^2 y}{dt^2} = -k|v|^{n-1}v_y - g,
\end{align}

where $\vec r$ is the position vector and $v = \sqrt{v_x^2+v_y^2}$ is the magnitude of velocity. <br>
    Since $\frac{dx}{dt} = v_x$ and $\frac{dy}{dt} = v_y$, the equations of motion can be written as first-order differential components in velocity components as
\begin{align}
    \frac{dv_x}{dt} = -k|v|^{n-1}v_x \quad and \quad \frac{dv_y}{dt} = -k|v|^{n-1}v_y - g.
\end{align}


```python
import numpy as np
import matplotlib.pyplot as plt

# constants
g = 9.8
na, ka = 1, 0.8    # for n=1
dt = 0.01          # time step

# coefficient for frictional force for any n and initial velocity
def k_func(n, v0):
    return ka * v0**(na-n)

# magnitude of velocity
def v_func(vx, vy):
    return np.sqrt(vx**2 + vy**2)

# velocity components from magnitude of velocity and angle with the horizontal
def vx_func(v, theta):
    return v*np.cos(theta)

def vy_func(v, theta):
    return v*np.sin(theta)

# acceleration components; equations in part a) are used
def ax_func(vx, v, n, k):
    return -k * v**(n-1) * vx

def ay_func(vy, v, n, k):
    return -k * v**(n-1) * vy - g
```

### numerical solution using Euler-Richardson's method
The iterative scheme is as follows:
\begin{align}
    a_{x,\ i} &= a_x(v_{x,\ i}, v_i, n, k) \\ 
    a_{y,\ i} &= a_x(v_{y,\ i}, v_i, n, k) \\
    \\
    v_{x,\ mid} &= v_{x,\ i} + a_{x,\ i} \cdot \frac{\Delta t}{2} \\
    v_{y,\ mid} &= v_{x,\ i} + a_{y,\ i} \cdot \frac{\Delta t}{2} \\
    v_{mid} &= v(v_{x, mid}, v_{y, mid}) \\
    a_{x,\ mid} &= a_x(v_{x,\ mid}, v_{mid}, n, k) \\
    a_{y,\ mid} &= a_y(v_{y,\ mid}, v_{mid}, n, k) \\
    \\
    v_{x,\ i+1} &= v_{x,\ i} + a_{x,\ mid} \Delta t \\
    v_{y,\ i+1} &= v_{y,\ i} + a_{y,\ mid} \Delta t \\
    x_{i+1} &= x_i + v_{x,\ mid} \Delta t \\
    y_{i+1} &= y_i + v_{y,\ mid} \Delta t
\end{align}

The function below returns an array containing $t,\ x,\ y,\ v_x,\ v_y,\ a_x,\ a_y,\ v$ in order from the first to the last elements.

```python
def numerical(n, v0, theta0):
    k = k_func(n, v0)
    
    # initializing the temporary variables 
    t = 0
    x = y = 0 
    vx, vy = vx_func(v0, theta0), vy_func(v0, theta0)
    v = v0
    ax, ay = ax_func(vx, v, n, k), ay_func(vy, v, n, k)
    
    # list for the solution
    soln = [[], [], [], [], [], [], [], []]

    while y >= 0:
        # see the indexing above
        soln[0].append(t)
        soln[1].append(x)
        soln[2].append(y)
        soln[3].append(vx)
        soln[4].append(vy)
        soln[5].append(ax)
        soln[6].append(ay)  
        soln[7].append(v)
        
        # mid values
        vx_mid = vx + ax * dt/2
        vy_mid = vy + ay * dt/2
        v_mid = v_func(vx_mid, vy_mid)
        ax_mid = ax_func(vx_mid, v_mid, n, k)
        ay_mid = ay_func(vy_mid, v_mid, n, k)
        
        # next values
        t += dt
        vx += ax_mid * dt
        vy += ay_mid * dt
        v = v_func(vx, vy)
        x += vx_mid * dt
        y += vy_mid * dt
        ax = ax_func(vx, v, n, k)
        ay = ax_func(vy, v, n, k)
        
    return np.array(soln)
```

### part b)

```python
v0 = 22
theta0 = 34 / 180 * np.pi

# case a: n=1
a = numerical(1, v0, theta0)
xa, ya = a[1], a[2]

# plot of trajectory
plt.plot(xa, ya, label='$n=1$')
plt.title('trajectory for n=1')
plt.xlabel('$x\ [m]$')
plt.ylabel('$y\ [m]$')
plt.legend()
plt.show()
```

### part c)
The value of $k$ is already adjusted.

```python
# case b: b = 1.5
b = numerical(1.5, v0, theta0)
xb, yb = b[1], b[2]

# case c: c = 2
c = numerical(2, v0, theta0)
xc, yc = c[1], c[2]

# plot of trajectory for all cases
plt.plot(xa, ya, label='$n=1$')
plt.plot(xb, yb, label='$n=1.5$')
plt.plot(xc, yc, label='$n=2$')
plt.title('trajectory for all 3 cases')
plt.xlabel('$x\ [m]$')
plt.ylabel('$y\ [m]$')
plt.legend()
plt.show()
```

### part d)
The range and time aloft is the last element of the respective arrays for $x$ and $t$.

```python
# range
def R(soln):
    N = len(soln[1])
    return soln[1][N-1]

# time aloft
def T(soln):
    N = len(soln[0])
    return soln[0][N-1]

# variable v0; theta0 = 34 degrees
v_arr = np.array([2, 4, 8, 16, 32, 64, 128])
for i in range(len(v_arr)):
    v0 = v_arr[i]
    soln = numerical(1, v0, 34/180 * np.pi)
    
    plt.plot(soln[1], soln[2], label='v0= %i' %v0)

plt.title('trajectories for $\theta_0 =34^\circ$, n=1')
plt.legend()
plt.show()
    

num = 50
n_arr = np.linspace(1, 10, num)
R_arr = np.empty(num)
T_arr = np.empty(num)

for i in range(num):
    xx = numerical(n_arr[i], 100, 34/180*np.pi)
    R_arr[i] = R(xx)
    T_arr[i] = T(xx)
    
```
