```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special, integrate
```

### part a)
The gravitational force is approximated to a linear term and only the linear term in drag force is considered. <br>
$a=g(1-\frac{2y}{R})-kv$ <br>
The positive direction is taken to be vertically downwards. Given the initial values for velocity and position, the iteration scheme is as follows: <br> 
$a_n=g(1-\frac{2y_n}{R})-kv_n$ <br>
$v_{n+1} = v_n + a_n \Delta t$ <br>
$y_{n+1} = y_n + v_n \Delta t$ <br>

```python
# constants and assumptions (all values are in SI units)
g = 9.81
R = 6.37*10 ** 6
k = 1 # assumed

N = 1000 # number of iterations
t_initial, t_final = 0, 10
t = np.linspace(t_initial, t_final, N+1)
dt = (t_final - t_initial)/N # time step

a = np.empty(N+1, dtype=float)
v = np.empty(N+1, dtype=float)
y = np.empty(N+1, dtype=float)

# initialize
y[0] = 0
v[0] = 0
a[0] = g

# iteration
for i in range(1, N):
    y[i] = y[i-1] + v[i-1] * dt
    v[i] = v[i-1] + a[i-1] * dt
    a[i] = g*(1 - 2 * y[i] / R) - k * v[i]
    
# endpoint problem
y[N] = y[N-1]
v[N] = v[N-1]
a[N] = a[N-1]

# scatter plots of a, v, x vs t respectively
plt.figure(figsize=(15, 15))
plt.subplot(311)
plt.xlabel('t [s]')
plt.ylabel('a [m/s^2]')
plt.scatter(t, a, s=1)

plt.subplot(312)
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
plt.scatter(t, v, s=1)

plt.subplot(313)
plt.xlabel('t [s]')
plt.ylabel('y [m]')
plt.scatter(t, y, s=1)
plt.show()
```
