```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd


# random.randint(start, end) # both included
# random.randrange(start, stop[, step])

# constant parameters
p1, p2 = 0.0, 0.2   # dawdlen probabilities for two simulations 
v_max = 5           # maximum velocity
l = 100              # length of ring

# variable parameters
n = 18              # total no. of cars
a = 6               # initial no. of cars in traffic jam
b = n - a           # no. of cars initialized randomly
```

### initialize positions and velocities
- $a$ no. of cars are in traffic jam (positioned from 0 to $a-1$, with velocities 0)
- for remaining $n-a$ cars, distinct positions will be randomly generated and sorted
- then their velocities will be randomly generated, taking the car in front in consideration
- returns arrays of position and velocity

```python
def init_cars(n, a, p):
    b = n - a                 # randomly initialized cars
    s, v = np.empty(n, dtype=int), np.empty(n, dtype=int)
    
    for i in range(0, a):     # a cars in traffic jam
        s[i] = i
        v[i] = 0

    rand_pos = sorted(rnd.sample(range(a, l), b))  # sorted random positions for remaining cars
    
    for i in range(a, n):
        s[i] = rand_pos[i-a]

    for i in range(a, n):
        gap = (l + s[(i+1) % n] - s[i] - 1) % l  # periodic boundaries in car index and position 
        v_lim = gap if (gap < v_max) else v_max  # maximum possible velocity
        v[i] = rnd.randint(0, v_lim)         # random velocities
        
    return s, v
```

### Nagel-Schreckenberg model and spacetime plot
- t_max is the total no. of time steps 
- i is for time steps
- j represents jth car

```python
def NS(t_max, n, a, p, title=''):
    s = np.zeros([t_max, n], dtype=int)  # solutions for position and velocity for t_max iterations
    v = np.zeros([t_max, n], dtype=int)
    
    s[0:1, 0:n], v[0:1, 0:n] = init_cars(n, a, p)  # s, v here and in init_cars have different shapes
    
    # time iterations
    for i in range(1, t_max):
        # position in ith iteration = (position + velocity) % l in (i-1)th iteration
        s[i:i+1, 0:n] = (np.add(s[i-1:i, 0:n], v[i-1:i, 0:n])) % l 
        
        # velocity updates in ith iteration
        # it depends on velocity in (i-1)th iteration and distance in ith iteration
        for j in range(0, n):  # jth car
            d = (l + s[i, (j+1) % n] - s[i, j]) % l  # distance from next car; next car is at 0 position if j = n-1
            
            if (v[i-1, j] < v_max) and d > v[i-1, j] + 1:  # accelerate
                v[i, j] = v[i-1, j] + 1       
            elif d <= v[i-1, j]:                           # decelerate
                v[i, j] = d - 1
            else:                  # same velocity
                v[i, j] = v[i-1, j]
                
            if (rnd.random() < p) and v[i, j] > 0:  # implemented independently of other velocity updates
                v[i, j] -= 1

    plt.figure(figsize=(6, 6))
    for j in range(0, n):
        s_j = s[0:t_max, j:j+1]
        v_j = v[0:t_max, j:j+1]
        
        init = 0
        while init < t_max:
            t = init
            while (t+1 < t_max) and (s_j[t] <= s_j[t+1]):
                t += 1
            
            xx = s_j[init:t+1]
            yy = np.transpose(np.arange(init, t+1))
            plt.plot(xx, yy, 'b-', linewidth=1)
            
            init = t+1
        
    plt.title(title)
    plt.xlabel('position')
    plt.ylabel('time')
    plt.show()
```

### description and code verification
- each line represents a car (a car may have multiple lines if its position is reset)
- the velocity of a car is higher for almost horizontal lines and vice versa
- traffic jams are identified by a cluster of vertical lines
- the traffic jam moves towards the left since the cars behind will be affected

The three cases also serve as code verification. As expected, it can be seen that traffic jams increase with dawdle probability and car density. The fact that no two lines intersect each other suggests that the algorithm is properly implemented.

```python
# after the initial traffic jam, there is almost no jam
NS(100, n, a, p1, 'spacetime plot with $p=0$')
```

```python
# more traffic jams for the same car density due to dawdle probability
NS(100, n, a, p2, 'spacetime plot with $p=0.2$')
```

```python
# further increase in traffic jams due to higher car density
NS(100, 2*n, a, p2, 'spacetime plot with $p=0.2$ and double cars ')
```
